```python
import matplotlib.pyplot as plt
import json
```

```python
data = {
    "config_id": [1, 2, 3],
    "total_reward": [3.77, 9.7, 8.73],
    "success_rate": [0.54, 0.65, 0.65],
    "average_episode_length": [6.72, 7.21, 7.92]
}
```

```python
def create_bar_chart(x, y, title, xlabel, ylabel, filename):
    plt.figure(figsize=(8, 5))
    bars = plt.bar(x, y)

    # Adjust y-axis limits to add space above the tallest bar
    plt.ylim(0, max(y) * 1.2)  # 20% more space above the highest bar

    # Add labels above bars with dynamic positioning
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2.0, height + 0.05 * max(y),  # Offset based on max(y)
                 f"{height:.1f}", ha='center', fontsize=10)

    # Set chart title and labels
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # Save and display plot
    plt.savefig(filename)
    plt.show()
```

```python
create_bar_chart(
    data["config_id"],
    data["total_reward"],
    "Average Reward per Configuration",
    "Configuration ID",
    "Total Reward",
    "average_reward_per_config.png"
)
```

```python
# Plot Success Rate
create_bar_chart(
    data["config_id"],
    data["success_rate"],
    "Success Rate per Configuration",
    "Configuration ID",
    "Success Rate",
    "success_rate_per_config.png"
)
```

```python
# Plot Average Episode Length
create_bar_chart(
    data["config_id"],
    data["average_episode_length"],
    "Average Episode Length per Configuration",
    "Configuration ID",
    "Average Episode Length",
    "average_episode_length_per_config.png"
)
```

```python
training_data = {
    "timesteps": [9911, 49850, 99998, 499974],  # From shared logs and updates
    "average_reward": [-43.1, -40.8, -4.88, 8.05],  # ep_rew_mean over time
    "episode_length": [35.9, 34.4, 7.4, 7.67],  # ep_len_mean over time
    "exploration_rate": [0.05, 0.05, 0.1, 0.1],  # exploration_rate over time
    "loss": [0.272, 0.0111, 0.918, 3.79]  # loss values over time
}
```

```python
test_data = {
    "total_episodes": 300,  # Total test episodes based on logs.
    "successes": 95  # Episodes where the agent reached the treasure.
}
success_rate = (test_data["successes"] / test_data["total_episodes"]) * 100
print(f"Success Rate: {success_rate:.2f}%")  # Expected Output: Success Rate
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(training_data["timesteps"], training_data["average_reward"], label='Average Reward', marker='o')
plt.xlabel('Timesteps')
plt.ylabel('Average Reward')
plt.title('Reward Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(training_data["timesteps"], training_data["episode_length"], label='Episode Length', marker='o')
plt.xlabel('Timesteps')
plt.ylabel('Average Episode Length')
plt.title('Episode Length Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(len(training_data["loss"])), training_data["loss"], label='Loss', marker='o')
plt.xlabel('Training Iterations')
plt.ylabel('Loss')
plt.title('Loss Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(training_data["timesteps"], training_data["exploration_rate"], label='Exploration Rate', marker='o')
plt.xlabel('Timesteps')
plt.ylabel('Exploration Rate')
plt.title('Exploration Rate Over Time')
plt.legend()
plt.grid()
plt.show()
```
